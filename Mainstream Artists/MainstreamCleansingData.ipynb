```python
import pandas as pd
```

```python
file1 = './Mainstream DataSets/This_Is_Bad_Bunny.csv'
file2 = './Mainstream DataSets/This_Is_David_Guetta.csv'
file3 = './Mainstream DataSets/This_Is_Dua_Lipa.csv'
file4 = './Mainstream DataSets/This_Is_Ed_Sheeran.csv'
file5 = './Mainstream DataSets/This_Is_Harry_Styles.csv'
file6 = './Mainstream DataSets/This_Is_Justin_Bieber.csv'
file7 = './Mainstream DataSets/This_Is_Rihanna.csv'
file8 = './Mainstream DataSets/This_Is_Sam_Smith.csv'
file9 = './Mainstream DataSets/This_Is_Taylor_Swift.csv'
file10 = './Mainstream DataSets/This_Is_The_Weeknd.csv'
```

```python
df1 = pd.read_csv(file1, encoding="ISO-8859-1")
df2 = pd.read_csv(file2, encoding="ISO-8859-1")
df3 = pd.read_csv(file3, encoding="ISO-8859-1")
df4 = pd.read_csv(file4, encoding="ISO-8859-1")
df5 = pd.read_csv(file5, encoding="ISO-8859-1")
df6 = pd.read_csv(file6, encoding="ISO-8859-1")
df7 = pd.read_csv(file7, encoding="ISO-8859-1")
df8 = pd.read_csv(file8, encoding="ISO-8859-1")
df9 = pd.read_csv(file9, encoding="ISO-8859-1")
df10 = pd.read_csv(file10, encoding="ISO-8859-1")
```

```python
Mainstream_df = pd.concat([df1, df2, df3,df4,df5,df5,df6,df7,df8,df9,df10])
```

```python
Mainstream_df = Mainstream_df.rename(
    columns={"Unnamed: 0": "ID"})
```

```python
Mainstream_df.set_index(['ID'], inplace=True)
```

```python
Mainstream_df.drop_duplicates()
```

```python
Combined_Mainstream_df = Mainstream_df[Mainstream_df['artist'] == 'Bad Bunny' | Mainstream_df['artist'] == 'The Weeknd' ] 
```

```python
Combined_Mainstream_df.to_csv('./Mainstream DataSets/Combined Mainstream Dataset.csv', sep = ',')
```
