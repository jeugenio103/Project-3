```python
# Import libraries
import glob
import pandas as pd
```

```python
# Get CSV files list from a folder
path = './Mainstream DataSets/'
csv_files = glob.glob(path + "/*.csv")
```

```python
print(csv_files)
```

```python
# Read each CSV file into DataFrame
# This creates a list of dataframes
df_list = (pd.read_csv(file) for file in csv_files)

# Concatenate all DataFrames
big_df   = pd.concat(df_list,join="outer")
```

```python
big_df.head(1000)
```

```python
big_df.to_csv(f"./Mainstream Datasets/big_df.csv", sep = ',')
```
