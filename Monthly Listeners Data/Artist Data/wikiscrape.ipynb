```python
# import required modules
from bs4 import BeautifulSoup
import requests
import pandas as pd
 
# get URL
wiki_url = 'https://en.wikipedia.org/wiki/List_of_most-streamed_artists_on_Spotify'
table_class ="wikitable sortable jquery-tablesorter"

# scrape webpage
response = requests.get(wiki_url)
print(response.status_code)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
artist_table = soup.find('table',{'class':"wikitable"})
print(artist_table)
```

```python
# display scrapped data
df=pd.read_html(str(artist_table))
df=pd.DataFrame(df[0])
print(df.head())
```

```python
# drop the unwanted columns
data = df.drop(["Chg", "Ref(s)", "Unnamed: 5"], axis=1)
print(data)
```

```python
df.to_csv('top_spotify_artists.csv')
```
