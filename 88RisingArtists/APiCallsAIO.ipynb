```python
#Create function to retrieve each artist playlist ID
def allPlaylist(artists):
    #save playlist user name and ID 
    names = []
    #search for playlist
    results = sp.search(artists, type='playlist',limit='5')
    #append results of name and ID to list 
    names.append(results['playlists']['items'][0]['owner']['display_name'])
    names.append(results['playlists']['items'][0]['id'])
    
    return names
```

```python
#Create function to analyze each individual playlist track and add to DF 
def analyzePlaylist(spotify, playlist_ID):

    #Create column names for dataframe 
    list = ['song name','album','artist','popularity','featured',
                                'release_date','track_id',
                                'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
                                'instrumentalness', 'liveness', 'valence', 'tempo', 'duration_ms', 'time_signature']

    df = pd.DataFrame(columns = list)

    #Create empty dict for values 
    features = {}
   
    #Loop by searching through each track of playlist 
    playlist = sp.user_playlist_tracks(spotify,playlist_ID)['items']
    # artist = sp.artist(spotify,)
    for track in playlist:
        
        #Add meta results to dict 
        features['song name'] = track['track']['name']
        features['album'] = track['track']['album']['name']
        features['artist'] = track['track']['album']['artists'][0]['name']
        features['popularity'] = track['track']['popularity']
        for i in range(3):
            if i == 3: break
            try:
                features['featured'] = track['track']['artists'][i]['name']
            except:(IndexError)
            i += 1

        features['release_date'] = track['track']['album']['release_date']
        features['track_id'] = track['track']['id']
        # features['genres'] = track['track']['artists'][2]['genres'][0]

        #Loop by searching each audio feature & adding to dict 
        audio_features = sp.audio_features(features['track_id'])[0]
        for feature in list[8:]:
            features[feature] = audio_features[feature]

        #Create a single df row for each track 
        track_df = pd.DataFrame (features, index = [0] )
        #Concat previous row df to single df
        df = pd.concat([df, track_df], ignore_index = True)

    return df
```

```python
#Call function to sample single playlist
df = analyzePlaylist(spotify_id, playlist)
```

```python
#Create function to loop through each playlist ID and return single DF to concat to one DF
def analyzeAllPlaylists(playlist_dict):
    # Looop through each playlist ID (value)
    for i, (key, val) in enumerate(playlist_dict.items()):
        df = analyzePlaylist(*val)

        # Create counter to concat previous df
        if i == 0:
            concat_df = df
        else:
            concat_df = pd.concat([concat_df, df], ignore_index = True)
            
    return concat_df
```
